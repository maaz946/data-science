# data pre processing


```python
x=wine_datasets.drop('quality',axis=1)
```

```python
x
```

## label binerizing           for making true and false value

```python
y=wine_datasets['quality'].apply(lambda y_value: 1 if y_value >=7 else 0 )
```

```python
max(y)
```

```python
# lamda is used to replace values with 0 and 1
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
print(y.shape,y_train.shape,y_test.shape)
```

```python
# Random forrest algorithm
```

```python
model= RandomForestClassifier()
```

```python
model.fit(x_train,y_train)
```

## model Evaluation

### Accuracy score

```python
x_test_prdection=model.predict(x_test)
test_data_accuracy=accuracy_score(x_test_prdection,y_test)
print('Accuracy ',test_data_accuracy)
```
