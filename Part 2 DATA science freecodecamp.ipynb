```python
"""
Given a list of number i.e. [1,2,4,-5,7,9,3,2], make another list
that contains all the items in sorted order from min to max. i.e. your 
result will be another list like [-5,1,2,2,3,7,9]

"""
import numpy as np
L = [1,2,4,-5,7,9,3,2]
m =L[0]  # to find minimum number
idx = 0  #for finding index no
c = 0   #counter
for i in L:
    if i<m:
        m =i
        idx = c
    c+=1     #c is used for counting and we use another variable to represent idx to represent the c's no
print(idx,m)
```

```python
L = [1,2,4,-5,7,9,3,2]
for j in range(len(L)):
    m = L[j]
    idx = j
    c = j
    for i in range(j,len(L)):
        if L[i]<m:
            m = L[i]
            idx = c
        c+=1    
    tmp = L[j]
    L[j] = m
    L[idx] = tmp
print(L)
```

```python
"""Let say you are a teacher and you have different student
records containing id of a student and the marks list in each subject
where different students have taken different number of subjects. All
these records are in hard copy. You want to enter all the data in computer
and want to compute the average marks of each student and display"""

def forGettingMarksList():
    D={}
    while True:
        StudentId=input("enter Student ID")
        MarksList=input("Enter Marks by seperating Comma")
        MoreStudent=input("NO for stoping entering more Data entry")
        if StudentId in D:
            print("This id is already entered")
        else:
            D[StudentId]=MarksList.split(",")
        if MoreStudent.lower()=="no":
            return D
    
```

```python
def ForAverageMarks(D):
    AvegMarks={}
    for x in D:
        L=D[x]
        Sum=0
        for marks in L:
            Sum+=int(marks)
        AvegMarks[x]=Sum/len(L)
    return AvegMarks
```

```python
studentData=forGettingMarksList()
```
